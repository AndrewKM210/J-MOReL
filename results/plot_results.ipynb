```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import github_theme.config as git_config
```

```python
df_morel = pd.read_csv("logs/halfcheetah_medium_morel.csv")
df_morel
```

```python
df_jmorel = pd.read_csv("logs/halfcheetah_medium_jmorel.csv")
df_jmorel
```

```python
def plot_avg_return(dfs: [pd.DataFrame], 
                    metric: str, 
                    labels: [str], 
                    palette : [str], 
                    span : int = 25,
                    theme = 'dark', 
                    save_path : str = None):

    git_config.set_theme(theme)
    fig, ax = plt.subplots(figsize=git_config.FIGSIZE)

    for df, label, color in zip(dfs, labels, palette):
        df[f"{metric}_smooth"] = df[metric].ewm(span=span).mean()
        sns.lineplot(df, x="step", y=f"{metric}_smooth", c=color, label=label)

    ax.set_title("Average Return During Policy Search (halfcheetah-medium-v0)", fontsize=git_config.TITLE_FONTSIZE)
    ax.set_xlabel("Step")
    ax.set_ylabel("Average Return")
    ax.legend(title="")
    plt.tight_layout()
    plt.savefig(
            save_path, 
            dpi=git_config.DPI, 
            bbox_inches="tight", 
            transparent=False, 
            facecolor=plt.rcParams["figure.facecolor"])
    plt.show()

dfs = [df_morel, df_jmorel]
labels = ["MOReL", "J-MOReL"]
metric = "eval_score"
palette = [git_config.BASE_LIGHT, git_config.COLOR1_LIGHT]
plot_avg_return(dfs, metric, labels, palette, 25, 'light', '../assets/avg_return_light.png')
palette = [git_config.BASE_DARK, git_config.COLOR1_DARK]
plot_avg_return(dfs, metric, labels, palette, 25, 'dark', '../assets/avg_return_dark.png')
```
